```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python

df= pd.read_csv("Cab_Data.csv")
df.head(10)
```

```python
df['Date of Travel'] = pd.to_datetime(df['Date of Travel'], unit='D', origin='1899-12-30')
df.head()
```

```python
grouped_df = df.groupby(['Date of Travel', 'Company']).agg({'Price Charged': 'sum'}).reset_index()
```

```python
pivot_df = grouped_df.pivot_table(index='Date of Travel', columns='Company', values='Price Charged')
```

```python
plt.figure(figsize=(15, 7))
plt.plot(pivot_df.index, pivot_df['Pink Cab'], label='Pink Cab', color='magenta')
plt.plot(pivot_df.index, pivot_df['Yellow Cab'], label='Yellow Cab', color='yellow')
plt.xlabel('Date of Travel')
plt.ylabel('Transaction Amount')
plt.title('Transaction Amount by Company and Date')
plt.legend()
plt.show()
```

```python
df['Year'] = df['Date of Travel'].dt.year
df.head()
```

```python
yearly_grouped_df = df.groupby(['Year', 'Company']).agg({'Price Charged': 'sum'}).reset_index()
```

```python
import seaborn as sns

plt.figure(figsize=(9, 6))
sns.barplot(data=yearly_grouped_df, x='Year', y='Price Charged', hue='Company')
plt.xlabel('Year')
plt.ylabel('Total Price Charged')
plt.title('Total Price Charged by Each Company per Year')
plt.show()
```

```python
df['Margin'] = df['Price Charged'] - df['Cost of Trip']
df.head()
```

```python
date_grouped = df.groupby('Date of Travel').agg({'Margin': 'sum', 'Transaction ID': 'count'}).reset_index()
date_grouped = date_grouped.rename(columns={'Transaction ID': 'Num_Customers'})
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(date_grouped['Num_Customers'], date_grouped['Margin'])
plt.xlabel('Number of Customers')
plt.ylabel('Total Margin')
plt.title('Total Margin vs. Number of Customers')
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X = date_grouped['Num_Customers'].values.reshape(-1, 1)
```

```python
y = date_grouped['Margin']
```

```python
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
plt.figure(figsize=(12, 6))
plt.scatter(X, y, label='Actual Data')
plt.plot(X, y_pred, color='red', label='Fitted Line')
plt.xlabel('Number of Customers')
plt.ylabel('Total Margin')
plt.title('Total Margin vs. Number of Customers')
plt.legend()
plt.show()
```

```python
r_squared = model.score(X, y)
print(f"R-squared: {r_squared:.4f}")
```

```python
df_transc=pd.read_csv("Transaction_ID.csv")
```

```python
df_transc.head()
```

```python
merged_df = pd.merge(df, df_transc, left_on='Transaction ID', right_on='Transaction ID', how='inner')
merged_df.head()
```

```python
full_df = pd.merge(merged_df, df_customer, on='Customer ID', how='inner')
full_df.head()
```

```python
gender_stats = full_df.groupby('Gender').agg({'Age': 'mean', 'Income (USD/Month)': 'mean'}).reset_index()
print(gender_stats)
```

```python
# Set up the plotting area
fig, axes = plt.subplots(1, 2, figsize=(11, 6))

# Create the bar plots
sns.barplot(data=gender_stats, x='Gender', y='Age', ax=axes[0])
sns.barplot(data=gender_stats, x='Gender', y='Income (USD/Month)', ax=axes[1])

# Set the titles for the bar plots
axes[0].set_title('Average Age by Gender')
axes[1].set_title('Average Income by Gender')

# Display the plots
plt.show()
```

```python
yearly_profit = full_df.groupby(['Year', 'Company']).agg({'Margin': 'sum'}).reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_profit, x='Year', y='Margin', hue='Company', marker='o')
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.title('Yearly Profit Analysis of Cab Companies')
plt.show()
```

```python
full_df['Month'] = full_df['Date of Travel'].dt.month
monthly_profit = full_df.groupby(['Month', 'Company']).agg({'Margin': 'sum'}).reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_profit, x='Month', y='Margin', hue='Company', marker='o')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.title('Seasonal Profit Analysis of Cab Companies')
plt.show()
```

```python
def income_class(income):
    if income < 10000:
        return 'Low(<10000)'
    elif income < 20000:
        return 'Medium(10000-20000)'
    else:
        return 'High(20000<)'

full_df['Income Class'] = full_df['Income (USD/Month)'].apply(income_class)
```

```python
income_class_profit = full_df.groupby(['Company', 'Income Class']).agg({'Margin': 'sum', 'Customer ID': 'nunique'}).reset_index()
income_class_profit.rename(columns={'Customer ID': 'Customer Base'}, inplace=True)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Profit by income class plot
sns.barplot(data=income_class_profit, x='Income Class', y='Margin', hue='Company', ax=axes[0])
axes[0].set_title('Profit by Income Class for Cab Companies')

# Customer base by income class plot
sns.barplot(data=income_class_profit, x='Income Class', y='Customer Base', hue='Company', ax=axes[1])
axes[1].set_title('Customer Base by Income Class for Cab Companies')

plt.show()
```

```python
total_profit_by_company = income_class_profit.groupby('Company').agg({'Margin': 'sum'}).reset_index()
total_customers_by_company = income_class_profit.groupby('Company').agg({'Customer Base': 'sum'}).reset_index()
income_class_profit_normalized = income_class_profit.copy()
income_class_profit_normalized['Profit Percentage'] = income_class_profit_normalized.apply(lambda x: x['Margin'] / total_profit_by_company.loc[total_profit_by_company['Company'] == x['Company'], 'Margin'].values[0], axis=1)
income_class_profit_normalized['Customer Base Percentage'] = income_class_profit_normalized.apply(lambda x: x['Customer Base'] / total_customers_by_company.loc[total_customers_by_company['Company'] == x['Company'], 'Customer Base'].values[0], axis=1)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Profit percentage by income class plot
sns.barplot(data=income_class_profit_normalized, x='Income Class', y='Profit Percentage', hue='Company', ax=axes[0])
axes[0].set_title('Profit Percentage by Income Class for Cab Companies')

# Customer base percentage by income class plot
sns.barplot(data=income_class_profit_normalized, x='Income Class', y='Customer Base Percentage', hue='Company', ax=axes[1])
axes[1].set_title('Customer Base Percentage by Income Class for Cab Companies')

plt.show()
```
