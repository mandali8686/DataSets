```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python

df= pd.read_csv("Cab_Data.csv")
df.head(10)
```

```python
df['Date of Travel'] = pd.to_datetime(df['Date of Travel'], unit='D', origin='1899-12-30')
df.head()
```

```python
grouped_df = df.groupby(['Date of Travel', 'Company']).agg({'Price Charged': 'sum'}).reset_index()
```

```python
pivot_df = grouped_df.pivot_table(index='Date of Travel', columns='Company', values='Price Charged')
```

```python
plt.figure(figsize=(15, 7))
plt.plot(pivot_df.index, pivot_df['Pink Cab'], label='Pink Cab', color='magenta')
plt.plot(pivot_df.index, pivot_df['Yellow Cab'], label='Yellow Cab', color='yellow')
plt.xlabel('Date of Travel')
plt.ylabel('Transaction Amount')
plt.title('Transaction Amount by Company and Date')
plt.legend()
plt.show()
```

```python
df['Year'] = df['Date of Travel'].dt.year
df.head()
```

```python
yearly_grouped_df = df.groupby(['Year', 'Company']).agg({'Price Charged': 'sum'}).reset_index()
```

```python
import seaborn as sns

plt.figure(figsize=(9, 6))
sns.barplot(data=yearly_grouped_df, x='Year', y='Price Charged', hue='Company')
plt.xlabel('Year')
plt.ylabel('Total Price Charged')
plt.title('Total Price Charged by Each Company per Year')
plt.show()
```

```python
df['Margin'] = df['Price Charged'] - df['Cost of Trip']
df.head()
```

```python
date_grouped = df.groupby('Date of Travel').agg({'Margin': 'sum', 'Transaction ID': 'count'}).reset_index()
date_grouped = date_grouped.rename(columns={'Transaction ID': 'Num_Customers'})
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(date_grouped['Num_Customers'], date_grouped['Margin'])
plt.xlabel('Number of Customers')
plt.ylabel('Total Margin')
plt.title('Total Margin vs. Number of Customers')
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
X = date_grouped['Num_Customers'].values.reshape(-1, 1)
```

```python
y = date_grouped['Margin']
```

```python
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
plt.figure(figsize=(12, 6))
plt.scatter(X, y, label='Actual Data')
plt.plot(X, y_pred, color='red', label='Fitted Line')
plt.xlabel('Number of Customers')
plt.ylabel('Total Margin')
plt.title('Total Margin vs. Number of Customers')
plt.legend()
plt.show()
```

```python
r_squared = model.score(X, y)
print(f"R-squared: {r_squared:.4f}")
```

```python
yearly_profit = full_df.groupby(['Year', 'Company']).agg({'Margin': 'sum'}).reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_profit, x='Year', y='Margin', hue='Company', marker='o')
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.title('Yearly Profit Analysis of Cab Companies')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
ax = sns.lineplot(data=yearly_profit, x='Year', y='Margin', hue='Company', marker='o')
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.title('Yearly Profit Analysis of Cab Companies')

# Loop through the companies
for company in yearly_profit['Company'].unique():
    company_data = yearly_profit[yearly_profit['Company'] == company]
    prev_margin = None
    prev_year = None

    # Loop through the data points for each company
    for index, row in company_data.iterrows():
        year = row['Year']
        margin = row['Margin']

        # Add the margin value as text to the plot
        ax.text(year, margin, f"{margin:.2f}", horizontalalignment='left', size='small', color='black', weight='semibold')

        # Calculate the percentage increase or decrease
        if prev_margin is not None and prev_year is not None:
            percentage_change = (margin - prev_margin) / prev_margin * 100

            # Calculate the midpoint of the line segment
            midpoint_x = (year + prev_year) / 2
            midpoint_y = (margin + prev_margin) / 2

            # Add the percentage change text at the midpoint
            ax.text(midpoint_x, midpoint_y, f"({percentage_change:+.2f}%)", horizontalalignment='center', size='small', color='red' if percentage_change < 0 else 'green', weight='semibold')

        prev_margin = margin
        prev_year = year

plt.show()
```

```python
full_df['Month'] = full_df['Date of Travel'].dt.month
monthly_profit = full_df.groupby(['Month', 'Company']).agg({'Margin': 'sum'}).reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_profit, x='Month', y='Margin', hue='Company', marker='o')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.title('Seasonal Profit Analysis of Cab Companies')
plt.show()
```

```python
profit_df = full_df.groupby(['Company', 'Year', 'Month']).agg({'Margin': 'sum'}).reset_index()
pink_cab_profit = profit_df[profit_df['Company'] == 'Pink Cab']
yellow_cab_profit = profit_df[profit_df['Company'] == 'Yellow Cab']
```

```python
from sklearn.preprocessing import PolynomialFeatures

degree = 3
poly = PolynomialFeatures(degree)
months_passed = profit_df['Year'] * 12 + profit_df['Month']
months_passed_poly = poly.fit_transform(months_passed.values.reshape(-1, 1))
```

```python
from sklearn.linear_model import LinearRegression

pink_cab_reg = LinearRegression()
pink_cab_reg.fit(months_passed_poly[pink_cab_profit.index], pink_cab_profit['Margin'])

yellow_cab_reg = LinearRegression()
yellow_cab_reg.fit(months_passed_poly[yellow_cab_profit.index], yellow_cab_profit['Margin'])
```

```python
last_month = max(months_passed)
prediction_months = np.arange(last_month + 1, last_month + 13).reshape(-1, 1)
prediction_months_poly = poly.transform(prediction_months)

pink_cab_predictions = pink_cab_reg.predict(prediction_months_poly)

yellow_cab_predictions = yellow_cab_reg.predict(prediction_months_poly)
```

```python

pink_cab_profit_predicted = pink_cab_profit.append(pd.DataFrame({'Company': 'Pink Cab', 'Year': (prediction_months // 12).flatten(), 'Month': (prediction_months % 12).flatten(), 'Margin': pink_cab_predictions}), ignore_index=True)
yellow_cab_profit_predicted = yellow_cab_profit.append(pd.DataFrame({'Company': 'Yellow Cab', 'Year': (prediction_months // 12).flatten(), 'Month': (prediction_months % 12).flatten(), 'Margin': yellow_cab_predictions}), ignore_index=True)

plt.figure(figsize=(14, 6))
plt.plot(months_passed[pink_cab_profit.index], pink_cab_profit['Margin'], 'o', label='Pink Cab (Original)')
plt.plot(pink_cab_profit_predicted['Year'] * 12 + pink_cab_profit_predicted['Month'], pink_cab_profit_predicted['Margin'], label='Pink Cab (Prediction)')
plt.plot(months_passed[yellow_cab_profit.index], yellow_cab_profit['Margin'], 'o', label='Yellow Cab (Original)')
plt.plot(yellow_cab_profit_predicted['Year'] * 12 + yellow_cab_profit_predicted['Month'], yellow_cab_profit_predicted['Margin'], label='Yellow Cab (Prediction)')
plt.xlabel('Months Passed')
plt.ylabel('Profit')
plt.title('Seasonal Profit Analysis with Polynomial Regression Predictions')
plt.legend()
plt.show()
```

```python
def income_class(income):
    if income < 10000:
        return 'Low(<10000)'
    elif income < 20000:
        return 'Medium(10000-20000)'
    else:
        return 'High(20000<)'

full_df['Income Class'] = full_df['Income (USD/Month)'].apply(income_class)
```

```python
income_class_profit = full_df.groupby(['Company', 'Income Class']).agg({'Margin': 'sum', 'Customer ID': 'nunique'}).reset_index()
income_class_profit.rename(columns={'Customer ID': 'Customer Base'}, inplace=True)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Profit by income class plot
sns.barplot(data=income_class_profit, x='Income Class', y='Margin', hue='Company', ax=axes[0])
axes[0].set_title('Profit by Income Class for Cab Companies')

# Customer base by income class plot
sns.barplot(data=income_class_profit, x='Income Class', y='Customer Base', hue='Company', ax=axes[1])
axes[1].set_title('Customer Base by Income Class for Cab Companies')

plt.show()
```

```python
total_profit_by_company = income_class_profit.groupby('Company').agg({'Margin': 'sum'}).reset_index()
total_customers_by_company = income_class_profit.groupby('Company').agg({'Customer Base': 'sum'}).reset_index()
income_class_profit_normalized = income_class_profit.copy()
income_class_profit_normalized['Profit Percentage'] = income_class_profit_normalized.apply(lambda x: x['Margin'] / total_profit_by_company.loc[total_profit_by_company['Company'] == x['Company'], 'Margin'].values[0], axis=1)
income_class_profit_normalized['Customer Base Percentage'] = income_class_profit_normalized.apply(lambda x: x['Customer Base'] / total_customers_by_company.loc[total_customers_by_company['Company'] == x['Company'], 'Customer Base'].values[0], axis=1)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Profit percentage by income class plot
sns.barplot(data=income_class_profit_normalized, x='Income Class', y='Profit Percentage', hue='Company', ax=axes[0])
axes[0].set_title('Profit Percentage by Income Class for Cab Companies')

# Customer base percentage by income class plot
sns.barplot(data=income_class_profit_normalized, x='Income Class', y='Customer Base Percentage', hue='Company', ax=axes[1])
axes[1].set_title('Customer Base Percentage by Income Class for Cab Companies')

plt.show()
```

```python
def age_group(age):
    if age < 20:
        return 'Below 20'
    elif age < 30:
        return '20-29'
    elif age < 40:
        return '30-39'
    elif age < 50:
        return '40-49'
    else:
        return 'Above 50'

full_df['Age Group'] = full_df['Age'].apply(age_group)
```

```python
from sklearn.linear_model import LinearRegression

# Aggregate data by year and company
yearly_profit = full_df.groupby(['Year', 'Company'])['Margin'].sum().reset_index()

# Pivot the data for easier processing
pivot_yearly_profit = yearly_profit.pivot_table(index='Year', columns='Company', values='Margin').reset_index()

# Prepare data for linear regression
x = pivot_yearly_profit['Year'].values.reshape(-1, 1)
y_pink = pivot_yearly_profit['Pink Cab'].values
y_yellow = pivot_yearly_profit['Yellow Cab'].values

# Fit linear regression models
reg_pink = LinearRegression().fit(x, y_pink)
reg_yellow = LinearRegression().fit(x, y_yellow)

# Forecast profit for the next year (2019)
next_year = np.array([2019]).reshape(-1, 1)
predicted_pink_profit = reg_pink.predict(next_year)[0]
predicted_yellow_profit = reg_yellow.predict(next_year)[0]

print("Predicted Pink Cab profit for 2019: ${:,.2f}".format(predicted_pink_profit))
print("Predicted Yellow Cab profit for 2019: ${:,.2f}".format(predicted_yellow_profit))
```

```python

# Add the predicted profit to the pivot table
predicted_profit = pd.DataFrame({'Year': [2019], 'Pink Cab': [predicted_pink_profit], 'Yellow Cab': [predicted_yellow_profit]})
pivot_yearly_profit_extended = pivot_yearly_profit.append(predicted_profit, ignore_index=True)

# Plot the historical profit and predicted profit
plt.figure(figsize=(12, 6))
sns.lineplot(data=pivot_yearly_profit_extended, x='Year', y='Pink Cab', marker='o', label='Pink Cab')
sns.lineplot(data=pivot_yearly_profit_extended, x='Year', y='Yellow Cab', marker='o', label='Yellow Cab')

# Highlight the predicted profit point
plt.scatter(2019, predicted_pink_profit, marker='o', color='r', zorder=10, label='Predicted Pink Cab Profit')
plt.scatter(2019, predicted_yellow_profit, marker='o', color='b', zorder=10, label='Predicted Yellow Cab Profit')

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.title('Yearly Profit Analysis and Forecast for Cab Companies')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=pivot_yearly_profit_extended, x='Year', y='Pink Cab', marker='o', label='Pink Cab')
sns.lineplot(data=pivot_yearly_profit_extended, x='Year', y='Yellow Cab', marker='o', label='Yellow Cab')

# Highlight the predicted profit point
plt.scatter(2019, predicted_pink_profit, marker='o', color='r', zorder=10, label='Predicted Pink Cab Profit')
plt.scatter(2019, predicted_yellow_profit, marker='o', color='b', zorder=10, label='Predicted Yellow Cab Profit')

# Add y values to the vertices
for idx, row in pivot_yearly_profit_extended.iterrows():
    plt.text(row['Year'], row['Pink Cab'], f"{row['Pink Cab']:.0f}", ha='center', va='bottom', fontsize=9)
    plt.text(row['Year'], row['Yellow Cab'], f"{row['Yellow Cab']:.0f}", ha='center', va='bottom', fontsize=9)

# Calculate and display percentage increase/decrease on the lines
for company in ['Pink Cab', 'Yellow Cab']:
    for i in range(1, len(pivot_yearly_profit_extended)):
        profit_diff = pivot_yearly_profit_extended.loc[i, company] - pivot_yearly_profit_extended.loc[i - 1, company]
        percentage_change = (profit_diff / pivot_yearly_profit_extended.loc[i - 1, company]) * 100
        x_text = (pivot_yearly_profit_extended.loc[i - 1, 'Year'] + pivot_yearly_profit_extended.loc[i, 'Year']) / 2
        y_text = (pivot_yearly_profit_extended.loc[i - 1, company] + pivot_yearly_profit_extended.loc[i, company]) / 2
        plt.text(x_text, y_text, f"{percentage_change:.1f}%", ha='center', va='center', fontsize=9, bbox=dict(facecolor='white', edgecolor='none', alpha=0.5))

# Customize the plot
plt.xlabel('Year')
plt.ylabel('Total Profit')
plt.title('Yearly Profit Analysis and Forecast for Cab Companies')
plt.legend()
plt.show()
```
